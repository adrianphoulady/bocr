```python
import os
os.environ["TOKENIZERS_PARALLELISM"] = "false"
```

```python
from IPython.display import Markdown
from bocr.backbones.qwen import extract_text

md_content = extract_text("sample1.jpg")
display(Markdown(md_content))
```

```python
from bocr import Config, ocr

config = Config(model_id="Qwen/Qwen2-VL-7B-Instruct", export_results=True, export_format="pdf", verbose=True)
input_files = ["sample2.pdf"]
results = ocr(input_files, config)
print(results)
```

```python
from IPython.display import display, Markdown
from pdf2image import convert_from_path

for path, texts in zip(input_files, results.values()):  
    images = convert_from_path(path)  
    
    for image, text in zip(images, texts):  
        display(image)
        display(Markdown(text))
```
